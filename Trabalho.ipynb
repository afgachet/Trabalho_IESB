```python
#Importando bibliotecas

import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
#Importando arquivo com a base de dados

mundo = pd.read_csv('countries of the world.csv', decimal=',')
```

```python
#Analisando os dados da base

mundo.info()
```

```python
mundo.head()
```

```python
mundo.describe()
```

```python
mundo.describe(include='all')
```

```python
#Tratando os dados
print('Dados faltantes:')
print(mundo.isnull().sum())
```

```python
#Agrupamento pelas regioes e media dos campos para preenchimento dos valores faltantes

mundo.groupby('Region')[['GDP ($ per capita)', 'Literacy (%)', 'Agriculture']].median()
```

```python
# Imputando novos valores aos campos sem valor

for coluna in mundo.columns.values:
    if mundo[coluna].isnull().sum() == 0:
        continue
    if coluna == 'Climate':
        novo_valor = mundo.groupby('Region')['Climate'].apply(lambda x: x.mode().max())
    else:
        novo_valor = mundo.groupby('Region')[coluna].median()
    for region in mundo['Region'].unique():
        mundo[coluna].loc[(mundo[coluna].isnull())&(mundo['Region']==region)] = novo_valor[region]
```

```python
print('Dados faltantes:')
print(mundo.isnull().sum())
```

```python
#Os 20 paises com maior PIB per capta
print ("Os 20 paises com maior PIB per capta")
mundo.sort_values('GDP ($ per capita)',ascending=False).head(20)
```

```python
fig, ax = plt.subplots(figsize=(16,6))
maiores_PIB = mundo.sort_values('GDP ($ per capita)',ascending=False).head(20)
mean = pd.DataFrame({'Country':['Média Mundial'], 'GDP ($ per capita)':[mundo['GDP ($ per capita)'].mean()]})
gdps = pd.concat([maiores_PIB[['Country','GDP ($ per capita)']],mean],ignore_index=False)
sns.barplot(x='Country', y='GDP ($ per capita)', data=gdps, palette='Set1')
ax.set_xlabel("País", labelpad=15)
ax.set_ylabel("PIB ($ per capta)", labelpad=30)
ax.xaxis.label.set_fontsize(16)
ax.yaxis.label.set_fontsize(16)
plt.xticks(rotation=90)
plt.show()
```

```python
#Os 20 paises com menor PIB per capta
print ("Os 20 paises com menor PIB per capta")
mundo.sort_values('GDP ($ per capita)',ascending=True).head(20)
```

```python
fig, ax = plt.subplots(figsize=(16,6))
menores_PIB = mundo.sort_values('GDP ($ per capita)',ascending=True).head(20)
mean = pd.DataFrame({'Country':['Média Mundial'], 'GDP ($ per capita)':[mundo['GDP ($ per capita)'].mean()]})
gdps = pd.concat([menores_PIB[['Country','GDP ($ per capita)']],mean],ignore_index=False)
sns.barplot(x='Country', y='GDP ($ per capita)', data=gdps, palette='Set1')
ax.set_xlabel("País", labelpad=15)
ax.set_ylabel("PIB ($ per capta)", labelpad=30)
ax.xaxis.label.set_fontsize(16)
ax.yaxis.label.set_fontsize(16)
plt.xticks(rotation=90)
plt.show()
```

# 
## Analisando os graficos abaixo fica evidente a relação direta entre o PIB per capta e as demais variaveis exploradas.

### PIB per capta X Telefones por 1000 hab. quanto maior o PIB maior a porcentagem de telefones por habitantes
### PIB per capta X Taxa de natalidade, quanto menor o PIB maior a taxa de natalidade
### PIB per capta X Taxa de mortalidade infantil, quanto menor o PIB maior a taxa mortalidade infantil
### PIB per capta X  Agricultura , quanto maior o PIB menos agricola
### PIB per capta X Industria , quanto maior o PIB economia mais industrializada


```python
#Correlações entre PIB per capta e demais variaveis

fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(20,20))
plt.subplots_adjust(hspace=0.4)

corr_to_gdp = pd.Series()
for col in mundo.columns.values[2:]:
    if ((col!='GDP ($ per capita)')&(col!='Climate')):
        corr_to_gdp[col] = mundo['GDP ($ per capita)'].corr(mundo[col])
abs_corr_to_gdp = corr_to_gdp.abs().sort_values(ascending=False)
corr_to_gdp = corr_to_gdp.loc[abs_corr_to_gdp.index]

for i in range(3):
    for j in range(3):
        sns.regplot(x=corr_to_gdp.index.values[i*3+j], y='GDP ($ per capita)', data=mundo,
                   ax=axes[i,j], fit_reg=False, marker='.')
        title = 'PIB_per_capta X '+str(col)
        axes[i,j].set_title(title)
axes[1,2].set_xlim(0,102)
plt.show()
```
